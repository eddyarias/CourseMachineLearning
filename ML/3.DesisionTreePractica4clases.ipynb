```python
import matplotlib.pyplot as plt

max_depths = range(1, 30)  # Profundidades de 1 a 12
train_aucs = []
val_aucs = []

for max_depth in max_depths:
    dt = DecisionTreeClassifier(max_depth=max_depth)
    dt.fit(X_train, y_train)

    # AUC para el conjunto de entrenamiento
    y_pred_train = dt.predict_proba(X_train)
    train_auc = roc_auc_score(y_train, y_pred_train, multi_class='ovr')
    train_aucs.append(train_auc)

    # AUC para el conjunto de validación
    y_pred_val = dt.predict_proba(X_val)
    val_auc = roc_auc_score(y_val, y_pred_val, multi_class='ovr')
    val_aucs.append(val_auc)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_aucs, label='AUC Train', marker='o')
plt.plot(max_depths, val_aucs, label='AUC Val', marker='o')
plt.title('AUC en función de la profundidad del árbol')
plt.xlabel('Profundidad máxima del árbol')
plt.ylabel('AUC')
plt.xticks(max_depths)
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, det_curve
import seaborn as sns

dt = DecisionTreeClassifier(max_depth=9)
dt.fit(X_train, y_train)

# Realizar predicciones
y_pred = dt.predict(X_val)


# AUC para el conjunto de entrenamiento
y_pred_train = dt.predict_proba(X_train)
train_auc = roc_auc_score(y_train, y_pred_train, multi_class='ovr')
print("Train Auc: ",train_auc)

# AUC para el conjunto de validación
y_pred_val = dt.predict_proba(X_val)
val_auc = roc_auc_score(y_val, y_pred_val, multi_class='ovr')
print("Val auc: ",val_auc)
# Crear la matriz de confusión
cm = confusion_matrix(y_val, y_pred)

# Graficar la matriz de confusión
plt.figure(figsize=(3, 3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)

# Personalizar el gráfico
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.show()
```

```python
# Obtener las probabilidades para cada clase
y_pred_proba = dt.predict_proba(X_val)

# Calcular ROC AUC para multiclase
roc_auc = roc_auc_score(y_val, y_pred_proba, multi_class='ovr')  # O 'multi_class='ovo''
print('ROC AUC:', roc_auc)

# Inicializar listas para FPR y FNR
fpr = []
fnr = []
n_classes = y_pred_proba.shape[1]

# Calcular FPR y FNR para cada clase
for i in range(n_classes):
    # Obtener la curva ROC para cada clase
    fpr_i, tpr_i, thresholds = roc_curve(y_val, y_pred_proba[:, i], pos_label=i)
    
    # Calcular FNR
    fnr_i = 1 - tpr_i
    
    # Almacenar FPR y FNR
    fpr.append(fpr_i)
    fnr.append(fnr_i)

# Graficar la curva DET
plt.figure(figsize=(8, 6))

for i in range(n_classes):
    plt.plot(fpr[i], fnr[i], marker='o', label=f'Clase {i}')

plt.title('Curva DET (Detection Error Tradeoff) por Clase')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Falsos Negativos (FNR)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--")

# Añadir línea de referencia (error igual)
plt.plot([0.001, 1], [0.001, 1], color='red', linestyle='--', label='Error Igual')

plt.legend()
plt.show()
```
